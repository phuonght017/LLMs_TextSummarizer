````python
def markdown_to_beautiful_brochure(markdown_text, output_path='beautiful_brochure.png', 
                                  width=800, height=1100, 
                                  theme='professional', 
                                  include_footer=True,
                                  company_logo=None):
    """
    Convert markdown text to a beautifully styled brochure image
    
    Args:
        markdown_text (str): The markdown text to convert
        output_path (str): Path to save the output image
        width (int): Width of the output image
        height (int): Height of the output image
        theme (str): Design theme ('professional', 'creative', 'minimal', 'elegant')
        include_footer (bool): Whether to include a footer with contact info
        company_logo (str): Path to company logo image (optional)
    """
    # Set theme-based styling
    themes = {
        'professional': {
            'bg_color': (245, 245, 250),
            'accent_color': (41, 128, 185),
            'text_color': (44, 62, 80),
            'header_bg': (41, 128, 185),
            'header_text': (255, 255, 255),
            'quote_bg': (236, 240, 241),
            'fonts': {
                'title': ('Arial', 42, 'bold'),
                'heading': ('Arial', 28, 'bold'),
                'subheading': ('Arial', 22, 'bold'),
                'body': ('Arial', 16, 'normal'),
            }
        },
        'creative': {
            'bg_color': (255, 255, 255),
            'accent_color': (231, 76, 60),
            'text_color': (52, 73, 94),
            'header_bg': (231, 76, 60),
            'header_text': (255, 255, 255),
            'quote_bg': (249, 231, 231),
            'fonts': {
                'title': ('Verdana', 44, 'bold'),
                'heading': ('Verdana', 30, 'bold'),
                'subheading': ('Verdana', 24, 'bold'),
                'body': ('Verdana', 16, 'normal'),
            }
        },
        'minimal': {
            'bg_color': (255, 255, 255),
            'accent_color': (52, 73, 94),
            'text_color': (52, 73, 94),
            'header_bg': (245, 245, 245),
            'header_text': (52, 73, 94),
            'quote_bg': (245, 245, 245),
            'fonts': {
                'title': ('Arial', 40, 'bold'),
                'heading': ('Arial', 26, 'bold'),
                'subheading': ('Arial', 20, 'bold'),
                'body': ('Arial', 16, 'normal'),
            }
        },
        'elegant': {
            'bg_color': (252, 252, 252),
            'accent_color': (142, 68, 173),
            'text_color': (44, 62, 80),
            'header_bg': (142, 68, 173),
            'header_text': (255, 255, 255),
            'quote_bg': (241, 231, 254),
            'fonts': {
                'title': ('Georgia', 42, 'bold'),
                'heading': ('Georgia', 28, 'bold'),
                'subheading': ('Georgia', 22, 'bold'),
                'body': ('Georgia', 16, 'normal'),
            }
        }
    }
    
    # Use selected theme or default to professional
    theme_data = themes.get(theme.lower(), themes['professional'])
    
    # Convert markdown to HTML
    html = markdown.markdown(markdown_text, extensions=['tables', 'fenced_code', 'nl2br'])
    
    # Parse HTML
    soup = BeautifulSoup(html, 'html.parser')
    
    # Create a blank image with background
    img = Image.new('RGB', (width, height), color=theme_data['bg_color'])
    draw = ImageDraw.Draw(img)
    
    # Add subtle background pattern or texture
    add_background_texture(img, theme)
    
    # Load fonts
    fonts = {}
    try:
        for font_type, font_specs in theme_data['fonts'].items():
            font_name, font_size, font_style = font_specs
            try:
                if font_style == 'bold':
                    fonts[font_type] = ImageFont.truetype(f"{font_name}.ttf", font_size)
                else:
                    fonts[font_type] = ImageFont.truetype(f"{font_name}.ttf", font_size)
            except:
                # Fallback to default if specific font not found
                fonts[font_type] = ImageFont.load_default()
    except Exception:
        # Ultimate fallback
        fonts = {
            'title': ImageFont.load_default(),
            'heading': ImageFont.load_default(),
            'subheading': ImageFont.load_default(),
            'body': ImageFont.load_default()
        }
    
    # Add a decorative header area
    header_height = 120
    draw.rectangle([0, 0, width, header_height], fill=theme_data['header_bg'])
    
    # Add decorative element to header
    add_header_decoration(img, width, header_height, theme_data)
    
    # If logo provided, add it to the header
    if company_logo and os.path.exists(company_logo):
        try:
            logo = Image.open(company_logo)
            logo_size = (100, 100)
            logo = logo.resize(logo_size, Image.LANCZOS)
            img.paste(logo, (width - logo_size[0] - 20, 10), logo if logo.mode == 'RGBA' else None)
        except Exception:
            pass # Silently handle logo issues

    # Extract title (h1) if present and place in header
    title = None
    h1_tags = soup.find_all('h1')
    if h1_tags:
        title = h1_tags[0].get_text().strip()
        # Remove title from soup so it's not processed again
        h1_tags[0].decompose()

    # Add title to header area
    if title:
        text_width = get_text_width(title, fonts['title'])
        draw.text((50, (header_height - fonts['title'].size) // 2), 
                 title, font=fonts['title'], fill=theme_data['header_text'])
    
    # Start position for content (after header)
    x, y = 50, header_height + 30
    
    # Add a subtle shadow line under header
    draw.line([(0, header_height), (width, header_height)], 
              fill=tuple(max(0, c-30) for c in theme_data['header_bg']), width=2)
    
    # Track content for layout decisions
    content_blocks = []
    
    # First pass: collect content blocks
    for element in soup.find_all(['h2', 'h3', 'p', 'ul', 'ol', 'li', 'blockquote']):
        if element.parent.name in ['ul', 'ol'] and element.name == 'li':
            continue  # Skip list items for now, we'll process them with their parent
            
        tag_name = element.name
        content = element.get_text().strip()
        
        # Skip empty elements
        if not content:
            continue
            
        # Add to content blocks
        content_blocks.append((tag_name, content, element))
    
    # Second pass: render content with improved layout
    max_width = width - 100  # Margins on both sides
    
    for idx, (tag_name, content, element) in enumerate(content_blocks):
        # Check if we need a new page
        if y > height - 150:  # Leave room for footer
            # In a real implementation, create a new page
            break
            
        # Handle different markdown elements with enhanced styling
        if tag_name == 'h2':
            # Add some space before headings
            y += 15
            
            # Add a small accent box before heading
            accent_width = 10
            accent_height = 30
            draw.rectangle([x - 20, y, x - 10, y + accent_height], 
                         fill=theme_data['accent_color'])
            
            # Heading text
            draw.text((x, y), content, font=fonts['heading'], fill=theme_data['accent_color'])
            y += fonts['heading'].size + 5
            
            # Add a line under h2
            draw.line([(x, y), (width-x, y)], fill=theme_data['accent_color'], width=2)
            y += 20
            
        elif tag_name == 'h3':
            y += 10
            draw.text((x, y), content, font=fonts['subheading'], fill=theme_data['text_color'])
            y += fonts['subheading'].size + 15
            
        elif tag_name == 'p':
            # Check if this is the first paragraph after a heading for special treatment
            is_first_after_heading = False
            if idx > 0 and content_blocks[idx-1][0] in ['h2', 'h3']:
                is_first_after_heading = True
                
            # Special style for first paragraph after heading
            if is_first_after_heading:
                # Slightly larger first letter (drop cap)
                first_char = content[0]
                rest_content = content[1:]
                
                # Draw the drop cap
                drop_cap_font = create_larger_font(fonts['body'], scale=1.5)
                draw.text((x, y), first_char, font=drop_cap_font, fill=theme_data['accent_color'])
                
                # Calculate offset for rest of text
                drop_cap_width = get_text_width(first_char, drop_cap_font)
                
                # Draw rest of the first line
                wrapped_lines = textwrap.wrap(rest_content, width=int((max_width-drop_cap_width-10)/10))
                
                if wrapped_lines:
                    draw.text((x + drop_cap_width + 5, y), wrapped_lines[0], 
                             font=fonts['body'], fill=theme_data['text_color'])
                    y += fonts['body'].size + 8
                    
                    # Draw remaining lines
                    for line in wrapped_lines[1:]:
                        draw.text((x, y), line, font=fonts['body'], fill=theme_data['text_color'])
                        y += fonts['body'].size + 8
                
            else:
                # Regular paragraph
                wrapped_text = textwrap.fill(content, width=int(max_width/10))
                for line in wrapped_text.splitlines():
                    draw.text((x, y), line, font=fonts['body'], fill=theme_data['text_color'])
                    y += fonts['body'].size + 8
                    
            y += 10  # Extra space after paragraph
            
        elif tag_name in ['ul', 'ol']:
            y += 10  # Space before list
            items = element.find_all('li')
            
            for i, item in enumerate(items):
                item_content = item.get_text().strip()
                
                # Add bullet or number with accent color
                prefix = "• "
                if tag_name == 'ol':
                    prefix = f"{i+1}. "
                
                draw.text((x, y), prefix, font=fonts['body'], fill=theme_data['accent_color'])
                prefix_width = get_text_width(prefix, fonts['body'])
                
                # Wrapped text for item
                wrapped_item = textwrap.fill(item_content, width=int((max_width-prefix_width-10)/10))
                first_line = True
                
                for line in wrapped_item.splitlines():
                    if first_line:
                        draw.text((x + prefix_width, y), line, font=fonts['body'], fill=theme_data['text_color'])
                        first_line = False
                    else:
                        # Indent continuation lines
                        draw.text((x + prefix_width + 10, y), line, font=fonts['body'], fill=theme_data['text_color'])
                        
                    y += fonts['body'].size + 8
                    
                y += 5  # Space between list items
                
            y += 10  # Space after list
            
        elif tag_name == 'blockquote':
            y += 15  # Space before quote
            
            # Create a nice background box for the quote
            quote_padding = 15
            
            # Calculate quote height
            wrapped_quote = textwrap.fill(content, width=int((max_width-50)/10))
            line_count = len(wrapped_quote.splitlines())
            quote_height = (line_count * (fonts['body'].size + 8)) + (quote_padding * 2)
            
            # Draw quote box with rounded corners
            quote_box = Image.new('RGBA', (max_width, quote_height), (0, 0, 0, 0))
            quote_draw = ImageDraw.Draw(quote_box)
            rounded_rectangle(quote_draw, (0, 0, max_width, quote_height), 
                             theme_data['quote_bg'], radius=10)
            
            # Add a vertical accent line for the quote
            quote_draw.rectangle([5, quote_padding, 8, quote_height - quote_padding], 
                                fill=theme_data['accent_color'])
            
            # Paste quote box onto main image
            img.paste(quote_box, (x - quote_padding, y), quote_box)
            
            # Add quote text
            quote_x = x + 15
            quote_y = y + quote_padding
            
            for line in wrapped_quote.splitlines():
                draw.text((quote_x, quote_y), line, font=fonts['body'], fill=theme_data['text_color'])
                quote_y += fonts['body'].size + 8
                
            y += quote_height + 15  # Move past quote box
            
        # Add spacing between elements
        if tag_name not in ['li']:
            y += 5
    
    # Add footer if requested
    if include_footer:
        add_footer(img, draw, width, height, theme_data, fonts['body'])
    
    # Add subtle border
    draw.rectangle([0, 0, width-1, height-1], outline=theme_data['accent_color'], width=1)
    
    # Final enhancements - apply subtle effects
    img = img.filter(ImageFilter.SMOOTH)
    
    # Save the image
    img.save(output_path, quality=95)
    return img

# Helper functions
def add_background_texture(img, theme):
    """Add subtle texture or pattern to the background based on theme"""
    width, height = img.size
    
    if theme == 'professional':
        # Subtle gradient
        for y in range(height):
            factor = y / height
            for x in range(width):
                r, g, b = img.getpixel((x, y))
                r = int(r * (1 - factor * 0.1))
                g = int(g * (1 - factor * 0.1))
                b = int(b * (1 - factor * 0.05))
                img.putpixel((x, y), (r, g, b))
    
    elif theme == 'creative':
        # Add some random dots
        draw = ImageDraw.Draw(img)
        for _ in range(100):
            x = random.randint(0, width-1)
            y = random.randint(0, height-1)
            size = random.randint(1, 3)
            opacity = random.randint(5, 30)
            draw.ellipse([x, y, x+size, y+size], fill=(255, 255, 255, opacity))
    
    elif theme == 'elegant':
        # Subtle vignette effect
        mask = Image.new('L', (width, height), 255)
        draw = ImageDraw.Draw(mask)
        for i in range(min(width, height) // 10):
            opacity = 255 - i * 5
            if opacity < 0:
                opacity = 0
            draw.rectangle([i, i, width-i, height-i], outline=opacity)
        img = ImageOps.expand(img, border=1, fill=(0, 0, 0))

def add_header_decoration(img, width, header_height, theme_data):
    """Add decorative elements to the header area"""
    draw = ImageDraw.Draw(img)
    
    # Add decorative shapes based on theme
    accent_color = theme_data['accent_color']
    lighter_accent = tuple(min(255, c+30) for c in accent_color)
    
    # Draw accent shapes
    for i in range(5):
        x_pos = width - 200 + (i * 30)
        y_pos = header_height - 30
        size = 20 - (i * 3)
        draw.rectangle([x_pos, y_pos, x_pos + size, y_pos + size], 
                     fill=lighter_accent, outline=accent_color)

def add_footer(img, draw, width, height, theme_data, font):
    """Add a professional footer to the brochure"""
    footer_height = 50
    footer_y = height - footer_height
    
    # Footer line
    draw.line([(50, footer_y), (width-50, footer_y)], 
             fill=theme_data['accent_color'], width=1)
    
    # Footer text
    footer_text = "Contact us for more information • www.example.com • info@example.com • (555) 123-4567"
    text_width = get_text_width(footer_text, font)
    
    draw.text(((width - text_width) // 2, footer_y + 15), 
             footer_text, font=font, fill=theme_data['text_color'])

def rounded_rectangle(draw, xy, color, radius=10):
    """Draw a rounded rectangle"""
    x1, y1, x2, y2 = xy
    draw.rectangle([x1+radius, y1, x2-radius, y2], fill=color)
    draw.rectangle([x1, y1+radius, x2, y2-radius], fill=color)
    draw.ellipse([x1, y1, x1+radius*2, y1+radius*2], fill=color)
    draw.ellipse([x2-radius*2, y1, x2, y1+radius*2], fill=color)
    draw.ellipse([x1, y2-radius*2, x1+radius*2, y2], fill=color)
    draw.ellipse([x2-radius*2, y2-radius*2, x2, y2], fill=color)

def get_text_width(text, font):
    """Get the width of text with the given font"""
    try:
        return font.getlength(text)
    except AttributeError:
        # Fallback for older PIL versions
        return font.getsize(text)[0]

def create_larger_font(original_font, scale=1.5):
    """Create a larger version of a font"""
    try:
        # Try to get font path and create larger version
        if hasattr(original_font, 'path'):
            return ImageFont.truetype(original_font.path, int(original_font.size * scale))
    except:
        pass
    
    # Fallback to default font
    return ImageFont.load_default()
````

````python
# A class to represent a Webpage

# Some websites need you to use proper headers when fetching them:
headers = {
 "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

class Website:
    """
    A utility class to represent a Website that we have scraped
    Attributes: url, body, title, text, links
    """

    def __init__(self, url):
        self.url = url
        response = requests.get(url, headers=headers)
        self.body = response.content
        soup = BeautifulSoup(self.body, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        if soup.body:
            for irrelevant in soup.body(["script", "style", "img", "input"]):
                irrelevant.decompose()
            self.text = soup.body.get_text(separator="\n", strip=True)
        else:
            self.text = ""
        links = [link.get('href') for link in soup.find_all('a')]
        self.links = [link for link in links if link]

    def get_contents(self):
        return f"Webpage Title:\n{self.title}\nWebpage Contents:\n{self.text}\n\n"
````

## First step: Have GPT-4o-mini figure out which links are relevant

### Use a call to gpt-4o-mini to read the links on a webpage, and respond in structured JSON.  
It should decide which links are relevant, and replace relative links such as "/about" with "https://company.com/about".  
We will use "one shot prompting" in which we provide an example of how it should respond in the prompt.

This is an excellent use case for an LLM, because it requires nuanced understanding. Imagine trying to code this without LLMs by parsing and analyzing the webpage - it would be very hard!


````python
"""
    Create system prompt, applying one-shot prompting
"""
link_system_prompt = "You are provided with a list of links found on a webpage. \
You are able to decide which of the links would be most relevant to include in a brochure about the company, \
such as links to an About page, or a Company page, or Careers/Jobs pages.\n"
link_system_prompt += "You should respond in JSON as in this example:"
link_system_prompt += """
{
    "links": [
        {"type": "about page", "url": "https://full.url/goes/here/about"},
        {"type": "careers page": "url": "https://another.full.url/careers"}
    ]
}
"""
````

````python
print(link_system_prompt)
````

````python
def get_links_user_prompt(website):
    user_prompt = f"Here is the list of links on the website of {website.url} - "
    user_prompt += "please decide which of these are relevant web links for a brochure about the company, respond with the full https URL in JSON format. \
Do not include Terms of Service, Privacy, email links.\n"
    user_prompt += "Links (some might be relative links):\n"
    user_prompt += "\n".join(website.links)
    return user_prompt
````

````python
def get_links(url):
    website = Website(url)
    response = openai.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": link_system_prompt},
            {"role": "user", "content": get_links_user_prompt(website)}
      ],
        response_format={"type": "json_object"}
    )
    result = response.choices[0].message.content
    return json.loads(result)
````

## Second step: make the brochure!

Assemble all the details into another prompt to GPT4-o

````python
def get_all_details(url):
    """
        Function get all content of the website and content from related refer links in that website
    """
    result = "Landing page:\n"
    result += Website(url).get_contents()
    links = get_links(url)
    print("Found links:", links)
    for link in links["links"]:
        result += f"\n\n{link['type']}\n"
        result += Website(link["url"]).get_contents()
    return result
````

````python
system_prompt = "You are an assistant that analyzes the contents of several relevant pages from a company website \
and creates a short brochure about the company for prospective customers, investors and recruits. Respond in markdown.\
Include details of company culture, customers and careers/jobs if you have the information."
````

````python
def get_brochure_user_prompt(company_name, url):
    user_prompt = f"You are looking at a company called: {company_name}\n"
    user_prompt += f"Here are the contents of its landing page and other relevant pages; use this information to build a short brochure of the company in markdown.\n"
    user_prompt += get_all_details(url)
    user_prompt = user_prompt[:5_000] # Truncate if more than 5,000 characters
    return user_prompt
````

````python
def create_brochure(company_name, url):
    response = openai.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": get_brochure_user_prompt(company_name, url)}
          ],
    )
    result = response.choices[0].message.content
    #display(Markdown(result))
    return result
````

## Using stream
Update the function to make results stream back from OpenAI,
with the familiar typewriter animation

````python
def stream_brochure(company_name, url):
    stream = openai.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": get_brochure_user_prompt(company_name, url)}
          ],
        stream=True
    )
    
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        response = response.replace("```","").replace("markdown", "")
        update_display(Markdown(response), display_id=display_handle.display_id)
````

````python
stream_brochure("HuggingFace", "https://huggingface.co")
````
